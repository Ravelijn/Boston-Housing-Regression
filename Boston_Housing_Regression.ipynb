```python
# STATISTICAL TEST
"""Q1 : Apakah ada perbedaan yang signifikan pada median dari rumah yang dibatasi oleh Sungai Charles dengan yang tidak dibatasi?"""

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
plt.figure(figsize=(10,4))
plt.title("Median value dari rumah yang dibatasi oleh Sungai Charles [1] atau tidak [0]")
sns.barplot(x='CHAS',y='MEDV',data=df)
```

```python
#T-Test for hypothesis testing
Ho = "Tidak Ada Perbedaaan Signifikan pada Median dari Kedua Sampel"   # Stating the Null Hypothesis
Ha = "Terdapat Perbedaan Signifikan Pada median dari kedua Sampel"   # Stating the Alternate Hypothesis

x = np.array(df[df.CHAS == 0].MEDV)
y = np.array(df[df.CHAS == 1].MEDV)
t, p_value  = stats.ttest_ind(x,y, axis = 0)  #Performing an Independent t-test

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')
```

```python
# STATISTICAL TEST
"""Q2 : Apakah ada perbedaan median dari rumah yang didiami dan dibangun sebelum tahun 1940 dengan yang sesudah tahun 1940??"""

#Discretizing the age variable into three groups of 35 years and younger, between 35 and 70 years and 70 years and older)
df.loc[(df['AGE']<=35),'age']='35 and younger'
df.loc[(df['AGE']>35)&(df['AGE']<70),'age']='between 35 and 70'
df.loc[df['AGE']>=70,'age']='70 and older'
df.head()
```

```python
#plotting
plt.figure(figsize=(16,6))
ax = sns.boxplot(x='MEDV', y='age', data=df)
plt.title("Median Value of Owner-Occupied Homes by AGE Group")
plt.show()
```

```python
#ANOVA for hypothesis testing
Ho = "Tidak Ada Perbedaaan Signifikan pada Median dari Kedua Sampel"   # Stating the Null Hypothesis
Ha = "Terdapat Perbedaaan Signifikan pada Median dari Kedua Sampel"   # Stating the Alternate Hypothesis

x = np.array(df[df.AGE == '35 and younger'].MEDV)
y = np.array(df[df.AGE == 'between 35 and 70'].MEDV)
z = np.array(df[df.AGE == '70 and older'].MEDV)
f, p_value  = stats.f_oneway(x,y,z, axis = 0)  #Performing an ANOVA


if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value}) > 0.05')
```

```python
#REGRESSION
"""Single Predictor"""
"""Dasar Pemilihan Awal Predictor adalah yang memiliki korelasi paling tinggi"""

"""Model 1 : Menggunakan Single Predictor RM"""

# scatterplot to visualize the relationship between RM and MEDV
plt.figure(figsize=(6, 6))
sns.scatterplot(x = 'RM', y = 'MEDV', data = df)
plt.show()

```

```python
import statsmodels.api as sm
import math
#the independant variable
X = df['RM']
#add an intercept (beta_0) to our model
X = sm.add_constant(X)
# Dependant variable
Y=df['MEDV']

model1 = sm.OLS(Y, X).fit()
predictions = model1.predict(X)
model1.summary()
```

```python
sns.regplot(data=df,y='MEDV',x='RM') #RM   =    average number of rooms per dwelling
plt.ylabel("MEDV")
plt.xlabel("RM")
```

```python
#Uji Asumsi 1 : Mean Residual Harus Nol
residuals = model1.resid
residuals.mean()
```

```python
#Uji Asumsi 2 : Tidak ada Heteroscedasticity
from statsmodels.stats.diagnostic import het_white
from statsmodels.compat import lzip
import statsmodels.stats.api as sms

name = ["F statistic", "p-value"]
test = sms.het_goldfeldquandt(Y, X)
lzip(name, test)
```

```python
#Uji Asumsi 3 : Linearitas
# predicted values
fitted = model1.fittedvalues

#sns.set_style("whitegrid")
sns.residplot(x = Y, y = residuals , color="lightblue", lowess=True) 
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Residual PLOT")
plt.show()
```

```python
#Uji Asumsi 4 : Normalitas

# Plot histogram of residuals
sns.histplot(residuals, kde=True)
```

```python
# Plot q-q plot of residuals
residuals = model1.resid

import pylab
import scipy.stats as stats

stats.probplot(residuals, dist="norm", plot=pylab)
plt.show()

```

```python
#Uji Asumsi 5 : Multikolinearitas
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices
#find design matrix for regression model using 'rating' as response variable 
y, X = dmatrices('MEDV ~ RM', data=df, return_type='dataframe')

#create DataFrame to hold VIF values
vif_df = pd.DataFrame()
vif_df['variable'] = X.columns 

#calculate VIF for each predictor variable 
vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

#view VIF for each predictor variable 
print(vif_df)
```
